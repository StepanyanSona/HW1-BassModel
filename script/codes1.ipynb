```python
import numpy as np
import pandas as pd
from pathlib import Path
import scipy.optimize as opt
import matplotlib.pyplot as plt
from Helper import *
import os
```

Reading the data

```python
file_direction = Path(os.getcwd())
data = file_direction.parent / "data" / "brazilTV.xlsx"
df = pd.read_excel(data, sheet_name="Data", skiprows=4,header=None)
```

Cleaning the data

```python
df = df.iloc[:, [1, 4]].rename(columns={df.columns[1]: 'Year', df.columns[4]: 'Count'})
df = df[df['Year'].apply(lambda x: str(x).isdigit())]
df['Year'] = df['Year'].astype(int)
df['Count'] = pd.to_numeric(df['Count'], errors='coerce')
df = df.dropna()
df
```

Plotting the dataset

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Count'], marker='o')
plt.title('LCD of TV Over Time')
plt.xlabel('Year')
plt.ylabel('LCD')
plt.grid(True)
#plt.savefig('lcd_tv_over_time.png', dpi=300)  
plt.show()
```

4. Estimate Bass Model parameters

```python
df['Year'] -= df['Year'].min()
params, _ = opt.curve_fit(bass_model, df['Year'], df['Count'], p0=[0.03, 0.38, 16000])
p, q, M = params
print(f"Estimated Parameters: p={p:.4f}, q={q:.4f}, M={M:.2f}")
```

5. Predict the diffusion of the innovation selected in step 1.

```python
time_periods = df['Year'].values
counts = df['Count'].values

years_future = np.arange(df['Year'].max() + 1, df['Year'].max() + 7)
predicted_future = bass_model(years_future - df['Year'].min(), p, q, M)
predicted = bass_model(time_periods, p, q, M)

plt.figure(figsize=(10, 6))

plt.scatter(df['Year'] + df['Year'].min(), df['Count'], label='Observed Data', color='blue')
plt.plot(df['Year'] + df['Year'].min(), predicted, label='Bass Model Prediction (Observed)', color='red')
plt.plot(years_future + df['Year'].min(), predicted_future, label='Future Prediction (Bass Model)', color='red', linestyle='--')

plt.xlabel('Years since 2009')
plt.ylabel('Adoption Count')
plt.legend()
plt.title('Innovation Diffusion Prediction using the Bass Model')
#plt.savefig('diffusion_pred.png', dpi=300)  
plt.show()
```

Plotting the cumulative adoption

```python
adopters_per_period = bass_model(time_periods, p, q, M)
cumulative_adopters = np.cumsum(adopters_per_period)

plt.plot(time_periods, cumulative_adopters, label='Cumulative Adoption', color='green')
plt.xlabel('Time (Years)')
plt.ylabel('Cumulative Number of Adopters')
plt.title('Cumulative Adoption of Innovation Over Time')
plt.legend()
plt.grid(True)
#plt.savefig('cumulative_ad.png', dpi=300)  
plt.show()
```

7. Estimate the number of adopters by period.

```python
new_adopters = []
cumulative_adopters = []

total_adopters = 0
for t in years_future:
    new = bass_model(t, p, q, M)
    total_adopters += new
    new_adopters.append(new)
    cumulative_adopters.append(total_adopters)

forecast_data = {
    "Year": years_future,
    "New Adopters": [round(na, 2) for na in new_adopters],
    "Cumulative Adopters": [round(ca, 2) for ca in cumulative_adopters]
}

df_forecast = pd.DataFrame(forecast_data)
print(df_forecast)
```
